## Load and View Data

```python
def load_data(csv_path:str, usecols:list=None) -> pd.DataFrame:
    data = pd.read_csv(csv_path)
    data.columns = data.columns.str.lower()
    
    if usecols:
        usecols += ["date"]
        data = data[set(usecols)]
    
    return data

def resample_data_freq(data:pd.DataFrame, freq:str, target_col) -> pd.DataFrame:
    data = data.copy()
    
    data.date = pd.to_datetime(data.date)
    data.set_index(["date"], inplace=True)
    data = data.resample(rule=freq).sum()
    
    data = data[data[target_col] != 0.0]
    
    data.reset_index(inplace=True)
    return data
    
    
target_col = "sales"
freq = "M"
csv_path = "train.csv"
lags = 13
test_ratio = 12


sales_data = load_data(csv_path=csv_path, usecols=[target_col])
freq_data = resample_data_freq(data=sales_data, target_col=target_col, freq=freq)

freq_data[target_col] = freq_data[target_col].astype(np.int64)
```

```python
# Duration of dataset
def sales_duration(data):
    data.date = pd.to_datetime(data.date)
    number_of_days = data.date.max() - data.date.min()
    number_of_years = round(number_of_days.days / 365)
    number_of_months = round(number_of_days.days / 30)
    numver_of_weeks = round(number_of_days.days / 7)
    
    print(number_of_days.days, 'days')
    print(numver_of_weeks, 'weeks')
    print(number_of_months, 'months')
    print(number_of_years, 'years')
    
sales_duration(data=sales_data)
```

```python
def sales_per_day(data, target_col, period):
    fig, ax = plt.subplots(figsize=(7,4))
    plt.hist(data[target_col], color='mediumblue')
    
    ax.set(xlabel = f"Sales Per {period}", ylabel = "Count", title = f"Distrobution of Sales Per {period.title()}")
    
sales_per_day(data=sales_data, target_col=target_col, period="day")
```

```python
def time_plot(data, x_col, y_col, title):
    fig, ax = plt.subplots(figsize=(15,5))
    sns.lineplot(x_col, y_col, data=data, ax=ax, color='mediumblue', label='Total Sales')
    
    if y_col.replace("_diff", "") == "total":
        second = data.groupby(data.date.dt.week)[y_col].mean().reset_index()
        second.date = pd.date_range(start=data.date.min(), end=data.date.max(), periods=13)
        sns.lineplot((second.date + datetime.timedelta(0*365/12)), y_col, data=second, ax=ax, color='red', label='Mean Sales')   
    else:
        second = data.groupby(data.date.dt.year)[y_col].mean().reset_index()
        second.date = pd.to_datetime(second.date, format='%Y')
        sns.lineplot((second.date + datetime.timedelta(6*365/12)), y_col, data=second, ax=ax, color='red', label='Mean Sales')   
    
    
    
    ax.set(xlabel = "Date", ylabel = f"{y_col.title()}", title = title)
    sns.despine()
    
time_plot(freq_data, 'date', target_col, 'Monthly Sales Before Diff Transformation')
```

```python
def plots(data, target_col, lags=None):
    # Convert dataframe to datetime index
    dt_data = data.set_index('date').drop(target_col, axis=1)
    dt_data.dropna(axis=0)
    
    layout = (1, 3)
    raw  = plt.subplot2grid(layout, (0, 0))
    acf  = plt.subplot2grid(layout, (0, 1))
    pacf = plt.subplot2grid(layout, (0, 2))
    
    dt_data.plot(ax=raw, figsize=(12, 5), color='mediumblue')
    smt.graphics.plot_acf(dt_data, lags=lags, ax=acf, color='mediumblue')
    smt.graphics.plot_pacf(dt_data, lags=lags, ax=pacf, color='mediumblue')
    sns.despine()
    plt.tight_layout()

plots(data=stationary_df, target_col=target_col, lags=lags);
```

Regressive modeling

```python
#create dataframe for transformation from time series to supervised
def generate_supervised(data, target_col, lags):
    supervised_df = data.copy()
    
    #create column for each lag
    for lag in range(1, lags):
        col_name = 'lag_' + str(lag)
        supervised_df[col_name] = supervised_df[f'{target_col}_diff'].shift(lag)
    
    #drop null values
    supervised_df = supervised_df.dropna().reset_index(drop=True)
    supervised_df.set_index(["date"], inplace=True)
    
    supervised_df.to_csv('model_df.csv')
    
    return supervised_df

model_df = generate_supervised(data=stationary_df, target_col=target_col, lags=lags)
model_df.head()
```

## Train Test Split

```python
test_length = int(len(model_df) * test_ratio) if test_ratio < 1 else test_ratio
```

```python
def tts(data: pd.DataFrame, target_col, test_length) -> tuple:
    data = data.copy()
    data.drop([target_col], axis=1, inplace=True)
        
    train, test = data[:-test_length].values, data[-test_length:].values
    
    return train, test

train, test = tts(model_df, target_col=target_col, test_length=test_length)
train.shape, test.shape
```

```python
def scale_data(train_set, test_set):
    #apply Min Max Scaler
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler = scaler.fit(train_set)
    
    # reshape training set
    train_set = train_set.reshape(train_set.shape[0], train_set.shape[1])
    train_set_scaled = scaler.transform(train_set)
    
    # reshape test set
    test_set = test_set.reshape(test_set.shape[0], test_set.shape[1])
    test_set_scaled = scaler.transform(test_set)
    
    X_train, y_train = train_set_scaled[:, 1:], train_set_scaled[:, 0:1].ravel()
    X_test, y_test = test_set_scaled[:, 1:], test_set_scaled[:, 0:1].ravel()
    
    return X_train, y_train, X_test, y_test, scaler

X_train, y_train, X_test, y_test, scaler_object = scale_data(train, test)
X_train.shape, X_test.shape
```

```python
def undo_scaling(y_pred, x_test, scaler_obj, lstm=False):  
    #reshape y_pred
    y_pred = y_pred.reshape(y_pred.shape[0], 1, 1)
    
    if not lstm:
        x_test = x_test.reshape(x_test.shape[0], 1, x_test.shape[1])
    
    #rebuild test set for inverse transform
    pred_test_set = []
    for index in range(0,len(y_pred)):
        pred_test_set.append(np.concatenate([y_pred[index], x_test[index]], axis=1))
        
    #reshape pred_test_set
    pred_test_set = np.array(pred_test_set)
    pred_test_set = pred_test_set.reshape(pred_test_set.shape[0], pred_test_set.shape[2])
    
    #inverse transform
    pred_test_set_inverted = scaler_obj.inverse_transform(pred_test_set)
    
    return pred_test_set_inverted
```

```python
def load_original_df(csv_path, freq):
    #load in original dataframe without scaling applied
    data = pd.read_csv(csv_path)
    data.columns = data.columns.str.lower()
    
    data.date = pd.to_datetime(data.date)
    data.set_index(["date"], inplace=True)
    data = data.resample(rule=freq).sum()
    
    data.reset_index(inplace=True)
    return data
```

```python
def predict_df(unscaled_predictions, original_df, target_col, test_length):
    #create dataframe that shows the predicted sales
    
    result_list = []
    sales_dates = list(original_df[-(test_length+1):].date)
    act_sales = list(original_df[-(test_length+1):][target_col])
    
    for index in range(0, len(unscaled_predictions)):
        result_dict = {}
        result_dict['pred_value'] = unscaled_predictions[index][0] + act_sales[index]
        result_dict['date'] = sales_dates[index+1]
        
        result_list.append(result_dict)
        
    df_result = pd.DataFrame(result_list)
    return df_result
```

```python
model_scores = {}

def get_scores(unscaled_df, original_df, target_col, test_length, model_name):
    
    rmse = np.sqrt(mean_squared_error(original_df[target_col][-test_length:], unscaled_df.pred_value[-test_length:]))
    mae = mean_absolute_error(original_df[target_col][-test_length:], unscaled_df.pred_value[-test_length:])
    
    r2 = r2_score(original_df[target_col][-test_length:], unscaled_df.pred_value[-test_length:])
    model_scores[model_name] = [rmse, mae, r2]

    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2 Score: {r2}")
```

```python
def plot_results(results, original_df, target_col, model_name):

    fig, ax = plt.subplots(figsize=(15,5))
    sns.lineplot(original_df.date, original_df[target_col], data=original_df, ax=ax, label='Original', color='mediumblue')
    sns.lineplot(results.date, results.pred_value, data=results, ax=ax, label='Predicted', color='Red')
    
    ax.set(xlabel="Date", ylabel=f"{target_col}", title=f"{model_name} {target_col.title()} Forecasting Prediction")
    
    ax.legend()
    sns.despine()
    
    plt.savefig(f'{model_name}_forecast.png')
```

```python
def run_model(train_data, test_data, model, model_name):
    
    X_train, y_train, X_test, y_test, scaler_object = scale_data(train_data, test_data)
    
    mod = model
    mod.fit(X_train, y_train)
    predictions = mod.predict(X_test)
    
    # Undo scaling to compare predictions against original data
#     original_df = load_original_df(csv_path=csv_path, freq=freq)
    original_df = freq_data
        
    unscaled = undo_scaling(predictions, X_test, scaler_object)
    unscaled_df = predict_df(unscaled, original_df, target_col=target_col, test_length=test_length)
    
    get_scores(unscaled_df, original_df, target_col=target_col, model_name=model_name, test_length=test_length)
    plot_results(unscaled_df, original_df, target_col=target_col, model_name=model_name)
    
    return unscaled_df, original_df
```

# LSTM MODEL

```python
def lstm_model(train_data, test_data, lstm_unit=4, epochs=200, batch_size=1, callbacks=None):
    X_train, y_train, X_test, y_test, scaler_object = scale_data(train_set=train, test_set=test_data)
    
    X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
    X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
    
    model = Sequential()
    model.add(LSTM(lstm_unit, batch_input_shape=(1, X_train.shape[1], X_train.shape[2]), stateful=True))
    model.add(Dense(1))
    model.add(Dense(1))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1, shuffle=False, callbacks=callbacks)
    
    predictions = model.predict(X_test,batch_size=1)
    
    original_df = freq_data
    unscaled = undo_scaling(predictions, X_test, scaler_object, lstm=True)
    unscaled_df = predict_df(unscaled, original_df, target_col=target_col, test_length=test_length)
    
    get_scores(unscaled_df, original_df, target_col=target_col, model_name='LSTM', test_length=test_length)
    plot_results(unscaled_df, original_df, target_col=target_col, model_name='LSTM')
    
lstm_model(train, test, lstm_unit=4, epochs=200, batch_size=1)

# dump all models scores
pickle.dump(model_scores, open("model_scores.p", "wb" ))
```

```python
def get_scores(data, target_col, test_length):
    
    model_scores = {}
    
    rmse = np.sqrt(mean_squared_error(data[target_col][-test_length:], data.forecast[-test_length:]))
    mae = mean_absolute_error(data[target_col][-test_length:], data.forecast[-test_length:])
    r2 = r2_score(data[target_col][-test_length:], data.forecast[-test_length:])
    model_scores['ARIMA'] = [rmse, mae, r2]
    
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2 Score: {r2}")
    
    pickle.dump(model_scores, open("arima_model_scores.p", "wb"))
```

```python
def sarimax_model(data, target_col):
    # Model
    sar = sm.tsa.statespace.SARIMAX(data[target_col], order=(12,0,0), seasonal_order=(0,1,0,12), trend='c').fit()

    # Predictions
    start, end, dynamic = 40, 100, 7
    data['forecast'] = sar.predict(start=start, end=end, dynamic=dynamic) 
    pred_df = data.forecast[start+dynamic:end]
    
    data[[target_col, 'forecast']].plot(color=['mediumblue', 'Red'])

    get_scores(data, target_col=target_col, test_length=test_length)

    return sar, data, pred_df

def predict_df(prediction_df, target_col, test_length):
    original_df = freq_data.reset_index()
        
    #create dataframe that shows the predicted sales
    result_list = []
    sales_dates = list(original_df[-(test_length+1):].date)
    act_sales = list(original_df[-(test_length+1):][target_col])
    
    for index in range(0, len(prediction_df)):
        result_dict = {}
        result_dict['pred_value'] = prediction_df[index] + act_sales[index]
        result_dict['date'] = sales_dates[index+1]
        result_list.append(result_dict)
        
    df_result = pd.DataFrame(result_list)
    
    return df_result, original_df

sar, ts_data, predictions = sarimax_model(datetime_df, target_col=target_col+"_diff")

prediction_df, original_df = predict_df(prediction_df=predictions, target_col=target_col, test_length=test_length)
plot_results(prediction_df, original_df, target_col=target_col, model_name="ARIMA")
```
