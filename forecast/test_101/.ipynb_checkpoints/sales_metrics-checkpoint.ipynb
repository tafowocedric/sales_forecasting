<h1 style="color: red;">Generate Sales Table For An AppstechLabUser</h1>

```python
def get_discount(sales_receipts):
    if sales_receipts[0].get('SalesItemLineDetail').get('DiscountAmt') == 0:
        if sales_receipts[1].get('DiscountLineDetail'):
            if sales_receipts[1].get('DiscountLineDetail').get('PercentBased'):
                return {
                    'discount_amt': sales_receipts[0].get('Amount') * (sales_receipts[1].get('DiscountLineDetail').get('DiscountPercent') / 100),
                    'discount_rate': sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate')
                }
            else:
                return {
                    'discount_amt': sales_receipts[0].get('SalesItemLineDetail').get('DiscountAmt'),
                    'discount_rate': sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate')
                }
        else:
            return {
                'discount_amt': sales_receipts[0].get('SalesItemLineDetail').get('DiscountAmt'),
                'discount_rate': sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate')
            }
    else:
        if sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate') == 0:
            return {
                'discount_amt': sales_receipts[0].get('SalesItemLineDetail').get('DiscountAmt'),
                'discount_rate': sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate')
            }
        else:
            return {
                'discount_amt': sales_receipts[0].get('Amount') * sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate'),
                'discount_rate': sales_receipts[0].get('SalesItemLineDetail').get('DiscountRate')
            }

    
def get_item_info(sales_receipts):
    return {'id': sales_receipts[0].get('SalesItemLineDetail').get('ItemRef').get('value'),
            'name': sales_receipts[0].get('SalesItemLineDetail').get('ItemRef').get('name'),
            'qty': sales_receipts[0].get('SalesItemLineDetail').get('Qty'),
            'unit_price': sales_receipts[0].get('SalesItemLineDetail').get('UnitPrice'),
            'amount': sales_receipts[0].get('Amount'),
            'tax_inclusive_amount': sales_receipts[0].get('SalesItemLineDetail').get('TaxInclusiveAmt'),
            'discount': get_discount(sales_receipts),
            'tax_ref_value': sales_receipts[0].get('SalesItemLineDetail').get('TaxCodeRef').get('value')
           }

def get_item_cost(item_id, appstech_labs_id):
    item_info = pd.read_sql_query(f"select purchase_cost  from quickbooks_items where id='{item_id}' and appstech_labs_id='{appstech_labs_id}'", cnx)
    return round(float(item_info['purchase_cost'][0]), 2)
    
def get_tax_code(_id, appstech_labs_id):
    
    tax_code_df = pd.read_sql_query(f"select * from quickbooks_tax_codes where id='{_id}' and appstech_labs_id='{appstech_labs_id}'", cnx)
    if tax_code_df['tax_group'][0] == True:
        if json.loads(tax_code_df['sales_tax_ratelist_nested'][0]):
            tax_rate_id = json.loads(tax_code_df['sales_tax_ratelist_nested'][0]).get('TaxRateDetail')[0].get('TaxRateRef').get('value')
            tax_rate_df = pd.read_sql_query(f"select * from quickbooks_tax_rates where id='{float(tax_rate_id)}' and appstech_labs_id='{appstech_labs_id}'", cnx)
            return round(float(tax_rate_df['rate_value'][0]), 2)
        else:
            return 0.0
    else:
        return 0.0
    
def get_customer_info(sales_receipts, index):
    return {
        'customer_id': sales_receipts['customer_ref_value'][index],
        'customer_name': sales_receipts['customer_ref_name'][index]
    }

def get_refunds(_id, date, customer_id):
    refund_receipts_df = pd.read_sql_query(f"select home_total_amt from quickbooks_refund_receipt where appstech_labs_id='{_id}' and txn_date='{date}' and customer_ref_value='{customer_id}'", cnx)
    if len(refund_receipts_df) > 0:
        return float(refund_receipts_df['home_total_amt'].sum())
    else:
        return float(0.0)
    
```

```python
def generate_sales_table(sales_receipts, appstech_labs_id):
    columns = ['date', 'item_id', 'item_name', 'qty', 'unit_price', 'cost', 'net_amount', 'net_cost', 'currency', 'discount_rate', 'discount_amount', 'sub_total', 'margin', 'tax_inclusive_amount', 'sales_tax_rate', 'sales_tax_amount', 'gross_amount', 'customer_id', 'customer_name', 'refund_total_amount']
    df = pd.DataFrame(columns=columns)
    
    for i in range(sales_receipts.shape[0]):
        customer_id = get_customer_info(sales_receipts, i)['customer_id']
        customer_name = get_customer_info(sales_receipts, i)['customer_name']
        date = sales_receipts['txn_date'][i]
        item_id = get_item_info(json.loads(sales_receipts['line_nested'][i]))['id']
        item_name = get_item_info(json.loads(sales_receipts['line_nested'][i]))['name']
        qty = get_item_info(json.loads(sales_receipts['line_nested'][i]))['qty']
        cost_price = get_item_cost(item_id, appstech_labs_id)
        unit_price = get_item_info(json.loads(sales_receipts['line_nested'][i]))['unit_price']
        net_amt = get_item_info(json.loads(sales_receipts['line_nested'][i]))['amount']
        net_cost = qty * cost_price
        currency_value = sales_receipts['currency_ref_value'][i]
        tax_inclusive_amt = get_item_info(json.loads(sales_receipts['line_nested'][i]))['tax_inclusive_amount']
        discount_rate = get_item_info(json.loads(sales_receipts['line_nested'][i]))['discount']['discount_rate']
        discount_amt = round(get_item_info(json.loads(sales_receipts['line_nested'][i]))['discount']['discount_amt'], 2)
        refunds_amount = get_refunds(appstech_labs_id, date, customer_id)
        sub_total = net_amt - discount_amt
        margin = sub_total - net_cost
        sales_tax_rate = get_tax_code(get_item_info(json.loads(sales_receipts['line_nested'][i]))['tax_ref_value'], appstech_labs_id)
        sales_tax_amount = sub_total * sales_tax_rate
        gross_amount = sub_total + sales_tax_amount - refunds_amount
                
        df = df.append({'date': date, 'item_id': item_id, 'item_name': item_name, 'qty': float(qty), 'unit_price': unit_price, 'cost': cost_price, 'net_amount': net_amt, 'net_cost': net_cost, 'currency': currency_value, 'tax_inclusive_amount': tax_inclusive_amt,
                  'discount_rate': float(discount_rate), 'discount_amount': float(discount_amt), 'sub_total': sub_total, 'margin': margin, 'sales_tax_rate': sales_tax_rate, 'sales_tax_amount': sales_tax_amount, 'gross_amount': gross_amount,
                  'customer_id': float(customer_id), 'customer_name': customer_name, 'refund_total_amount': refunds_amount}, ignore_index=True)
        
    return df
```

```python
apptech_labs_id = 2 # appstech labs user id
sales_receipts_df = pd.read_sql_query(f"select * from quickbooks_sales_receipts where appstech_labs_id='{apptech_labs_id}'", cnx)

sales_table_df = generate_sales_table(sales_receipts_df, apptech_labs_id)
sales_table_df.head()
```
